```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt

from scripts.mf_curved_sky import V_012_curved
from scripts.mf_flat_sky import V_012_flat
```

```python
# create a simple Gaussian random field healpy map (replace with your map)
nside = 256
lmax = 3 * nside

ell = np.arange(lmax + 1)
Cl = np.zeros_like(ell, dtype=float)
Cl[1:] = 1.0 / (ell[1:] * (ell[1:] + 1))

m = hp.synfast(Cl, nside=nside, lmax=lmax, new=True)

hp.mollview(m, cmap='viridis')
plt.show()
```

```python
# define thresholds
k_sigma = np.std(m)
min_sigma = -3
max_sigma = 3
threshold_count = 10

v = np.linspace(min_sigma * k_sigma, max_sigma * k_sigma, threshold_count)
```

```python
# calculate Minkowski functionals
V0, V1, V2 = V_012_curved(v, m)
```

```python
# plot Minkowski functionals
fig, axs = plt.subplots(1, 3, figsize=(18, 5.5))

axs[0].plot(V0, marker='.')
axs[0].set_title('$V_0$', fontsize=22)

axs[1].plot(V1, marker='.')
axs[1].set_title('$V_1$', fontsize=22)

axs[2].plot(V2, marker='.')
axs[2].set_title('$V_2$', fontsize=22)

plt.show()
```

```python
# --- Flat sky example ---

nside_flat = 256

m_square = np.random.normal(size=(nside_flat, nside_flat))

plt.figure(figsize=(6,6))
plt.imshow(m_square, origin='lower', cmap='viridis')
plt.title('Flat-sky Gaussian random field')
plt.colorbar()
plt.show()
```

```python
# define thresholds for flat sky map

k_sigma_flat = np.std(m_square)
min_sigma = -3
max_sigma = 3
threshold_count = 10

v_flat = np.linspace(min_sigma * k_sigma_flat, max_sigma * k_sigma_flat, threshold_count)
```

```python
# calculate Minkowski functionals (flat sky)

V0_flat, V1_flat, V2_flat = V_012_flat(v_flat, m_square)

# plot
fig, axs = plt.subplots(1, 3, figsize=(18, 5.5))

axs[0].plot(v_flat, V0_flat, marker='.')
axs[0].set_title('$V_0$ (flat sky)', fontsize=22)

axs[1].plot(v_flat, V1_flat, marker='.')
axs[1].set_title('$V_1$ (flat sky)', fontsize=22)

axs[2].plot(v_flat, V2_flat, marker='.')
axs[2].set_title('$V_2$ (flat sky)', fontsize=22)

plt.show()
```
