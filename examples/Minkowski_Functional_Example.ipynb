```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
import sys

sys.path.append("../scripts/")
from mf import V_012
```

## Curved Sky Example

```python
# load a convergence map
kappa_map = np.load("kappa_map_fullsky.npy")
hp.mollview(kappa_map, title='Kappa Map')
```

```python
# define thresholds (optional - this will be done automatically if you don't define v in the function)
k_sigma = np.std(kappa_map)
min_sigma = -3
max_sigma = 3
threshold_count = 20

v = np.linspace(min_sigma * k_sigma, max_sigma * k_sigma, threshold_count)
```

```python
# calculate Minkowski functionals
V0, V1, V2 = V_012(kappa_map, geometry="curved", v=v)
```

```python
# plot Minkowski functionals
fig, axs = plt.subplots(1, 3, figsize=(18, 5.5))
axs[0].plot(V0, marker='.')
axs[0].set_title('$V_0$', fontsize=22)
axs[1].plot(V1, marker='.')
axs[1].set_title('$V_1$', fontsize=22)
axs[2].plot(V2, marker='.')
axs[2].set_title('$V_2$', fontsize=22)
plt.show()
```

## Flat Sky Example

```python
# a simple Gaussian random field flat-sky convergence map (replace with your map)
nside = 256

kappa_square = np.random.normal(size=(nside, nside))

plt.imshow(kappa_square, origin='lower', cmap='viridis')
plt.show()
```

```python
# define thresholds
k_sigma = np.std(kappa_square)
min_sigma = -3
max_sigma = 3
threshold_count = 20

v = np.linspace(min_sigma * k_sigma, max_sigma * k_sigma, threshold_count)
```

```python
total_size_deg = 10.0 # adjust based on your map size
pixelsize = np.deg2rad(total_size_deg) / nside # this must be in radians

# calculate Minkowski functionals
V0, V1, V2 = V_012(m_square, geometry='flat', v=v, pixelsize=pixelsize)
```

```python
# plot Minkowski functionals
fig, axs = plt.subplots(1, 3, figsize=(18, 5.5))
axs[0].plot(V0, marker='.')
axs[0].set_title('$V_0$', fontsize=22)
axs[1].plot(V1, marker='.')
axs[1].set_title('$V_1$', fontsize=22)
axs[2].plot(V2, marker='.')
axs[2].set_title('$V_2$', fontsize=22)
plt.show()
```
